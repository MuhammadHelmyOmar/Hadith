## Large Corpus Creation Process (page 4)
- It is unbalanced for Mawdu and authentic Hadiths (authentic is the majority).
- Cleaning Phase:
  - Cleaned Matn and degree of authenticity columns.
  - For LK (34,088):
    - Using Dorar to check Hadiths when Arabic and English grades differ (2).
    - Removing Hadiths with empty Arabic Matn (826).
    - Removing Hadiths with empty English and Arabic grades (380).
    - Removing Hadiths with no English grade and an Arabic grade that we were not able to classify as authentic or not (32).
    - Keep the following grades:
      - Sahih - Authentic
      - Daif - Weak
      - Hassan - Good
      - Hassan - Sahih
      - Mawdu - Fabricated
      - Munkar
      - Remove the rest (708)
    - Remaining = 32,142
  - For NAH (7,363):
    - Scraping the degree from Hdith website for Hadiths without degree (3,552).
      - 900 Hadiths are detected as Mawdu
    - Removal of Hadiths without Matn (1,246).
    - Removal of authentic Hadiths (359).
    - Grouping Hadiths with degrees meaning Mawdu and leave the remaining as Daif, so grades:
      - Mawdu
      - Daif
    - Remaining = 5758
- NAHPlus contains 3660 Hadiths.
  - The enhanced version of NAH after cleaning and scraping.
  - Ony non-authentic
- Filter Phase:
  - Extract Mawdu from NAHPlus and LK(29).
  - Add 366 MHs scraped from Dorar.
  - Removing Duplicates.
  - MAHADDAT is the result and contains 2,452 MHs.
  - The final large corpus is a mix of:
    - MAHADDAT
    - SHs from LK
  - Duplicates are removed.
  - Hadiths with a Matn indicating their authenticity degree were identified and removed from the LK corpus.
  - Hadiths containing the words authentic (صحيح) or good (حسن) are removed.
  - Elemnate the starting expressions (قال رسول الله..إلخ) from LK.
- Final corpus contains 26,561 Hadiths.
  - 24,109 are Sahih
  - 2,452 are Mawdu
  - If we omit the extremes, the MHs have much lower words than the SHs.
  - For longer Hadiths, Mawdu ones can be three times longer than SHs, attaining a maximum of 5,608 words.
  - Built a list of stopwords specific to Hadiths, words like:
    - عز وجل: Almighty
    - الله: God
    - فقال: and he said
    - سمعت: I heard
    - النبي: The prophet
